```python
v = np.array([2, 4, 5, 6])
u = np.array([1, 0, 0, 2])
```

```python
def vector_vector_multiplication(u, v):

    assert v.shape[0] == u.shape[0]

    result = 0.0
    for i in range(u.shape[0]):

        result += u[i] * v[i]

    return result
```

```python
r = vector_vector_multiplication(u,v)
r
```

```python
# or use:

u.dot(v)
```

```python
U = np.array([
    [2, 4, 5, 6],
    [1, 2, 1, 2],
    [3, 1, 2, 1],
])
```

```python
U.shape
```

```python
def matrix_vector_multiplication(U, u):

    assert U.shape[1] == u.shape[0]
    
    num_rows = U.shape[0]
    result = np.zeros(num_rows)

    for i in range(num_rows):
        result[i] = vector_vector_multiplication(U[i], u)

    return result
```

```python
matrix_vector_multiplication(U, u)
```

```python
U.dot(u)
```

```python
V = np.array([
    [1, 1, 2],
    [0, 0.5, 1], 
    [0, 2, 1],
    [2, 1, 0],
])
```

```python
V[:, 0]
```

```python
def matrix_matrix_multiplication(U, V):

    assert U.shape[1] == V.shape[0]

    result = np.zeros((U.shape[0], V.shape[1]))

    for i in range(V.shape[1]):
        vi = V[:, i]
        Uvi = matrix_vector_multiplication(U, vi)
        result[:, i] = Uvi

    return result
```

```python
matrix_matrix_multiplication(U, V)
```

```python
U.dot(V)
```
