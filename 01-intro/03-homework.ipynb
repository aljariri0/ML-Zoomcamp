```python
condition = df['origin'] == 'Asia'

# columns = ['origin', 'model_year', 'vehicle_weight']
columns = ['model_year', 'vehicle_weight']

df.loc[condition, columns].head(7)
```

```python
X = df.loc[condition, columns].head(7).values
X
```

```python
XT = X.T
XTX = XT.dot(X)
```

```python
XTXI = np.linalg.inv(XTX)
```

```python
y = np.array([1100, 1300, 800, 900, 1000, 1100, 1200])
```

```python
w = XTXI.dot(XT).dot(y)
```

```python
w.sum().round(3)
```
